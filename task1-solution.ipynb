# Problem 1: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:
    Salary(Lakhs) : Tax(%)
        • Below 5 : 0%
        • 5-10 : 10%
        • 10-20 : 20%
        • aboove 20 : 30%

```python
# Program to calculate in-hand monthly salary from CTC
ctc = float(input("Enter your CTC in Lakhs: "))

# Convert lakhs to rupees
annual_salary = ctc * 100000

# Deductions
hra = 0.10 * annual_salary
da = 0.05 * annual_salary
pf = 0.03 * annual_salary

# Tax calculation based on slabs
if ctc < 5:
    tax_percent = 0
elif ctc <= 10:
    tax_percent = 10
elif ctc <= 20:
    tax_percent = 20
else:
    tax_percent = 30

tax_amount = (tax_percent / 100) * annual_salary

# In-hand annual salary
inhand_annual = annual_salary - (hra + da + pf + tax_amount)

# Monthly salary
monthly_salary = inhand_annual / 12

print("----- Salary Breakdown -----")
print(f"Annual Salary: ₹{annual_salary}")
print(f"HRA Deduction (10%): ₹{hra}")
print(f"DA Deduction (5%): ₹{da}")
print(f"PF Deduction (3%): ₹{pf}")
print(f"Tax ({tax_percent}%): ₹{tax_amount}")
print(f"In-hand Monthly Salary: ₹{monthly_salary}")
```

Problem 2: Write a program that take a user input of three angles and will find out whether it can form a triangle or not

```python
# Program to check if three angles form a triangle

a = float(input("Enter angle 1: "))
b = float(input("Enter angle 2: "))
c = float(input("Enter angle 3: "))

# Check triangle condition
if a > 0 and b > 0 and c > 0 and (a + b + c == 180):
    print("The angles form a valid triangle.")
else:
    print("The angles do NOT form a triangle.")
```
