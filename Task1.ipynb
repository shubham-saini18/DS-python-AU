## Problem 1
    Write a program that will calculate the in-hand monthly salary after deductions on CTC:
    HRA = 10%
    DA = 5%
    PF = 3%
    Taxes deduction:
    Below 5 Lakhs → 0%
    5–10 Lakhs → 10%
    10–20 Lakhs → 20%
    Above 20 Lakhs → 3

```python
ctc = float(input("Enter your CTC in lakhs : "))

annual_salary = ctc*100000

hra = 0.10 * annual_salary
da = 0.05 * annual_salary
pf = 0.03 * annual_salary 

if ctc <= 5:
    tax_percent = 0 
elif ctc <= 10:
    tax_percent = 10 
elif ctc <= 20:
    tax_percent = 20
else:
    tax_percent = 30

tax_amount = (tax_percent / 100) * annual_salary 

inhand_annual = annual_salary - (hra + da + pf + tax_amount)

monthly_salary = inhand_annual / 12

print("---- Salary Breakdown ----")
print(f"Annual salary : {annual_salary}")
print(f"HRA Deduction (10%) : {hra}")
print(f"DA Deduction (5%) : {da}")
print(f"PF Deduction (3%) : {pf}")
print(f"TAX ({tax_percent}% : {tax_amount})")
print(f"In Hand Monthly Salary : {monthly_salary}")
```

## Problem 3
    Write a program that takes user input of cost price and selling price and determines whether it is a loss or a profit.

```python
cp = float(input("Enter cost price: "))
sp = float(input("Enter selling price: "))

p = sp - cp
l = cp - sp

if sp > cp:
    print(f"It is a profit of {p} rs")
elif sp < cp:
    print(f"It is a loss of {l} rs")
else:
    print("No profit, no loss.")
```

## Problem 4: Menu Driven Program:
    cm to ft
    km to miles
    USD to INR

```python
while True:
    print("\n1. Convert cm to ft")
    print("2. Convert km to miles")
    print("3. Convert USD to INR")
    print("4. Exit")

    choice = input("Enter your choice (1/2/3/4): ")

    if choice == "1":
        cm = float(input("Enter value in cm: "))
        ft = cm / 30.48
        print("Feet:", ft)

    elif choice == "2":
        km = float(input("Enter value in km: "))
        miles = km * 0.621371
        print("Miles:", miles)

    elif choice == "3":
        usd = float(input("Enter amount in USD: "))
        inr = usd * 83.0     
        print("INR:", inr)

    elif choice == "4":
        print("Exit program")
        break

    else:
        print("Invalid choice")
```

## Problem 7 Reverse an Integer
    Example:
    Input: 76542
    Output: 24567

```python
num = int(input("Enter a number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

print("Reversed number:", rev)
```

## Problem 8 
    Take user input N.
    Compute the sum from 1 to N with conditions:
      Skip numbers divisible by 5
      Stop summing if total exceeds 300
     Do NOT use a for loop

    Example:
    Input: 30
    Output: 276

```python
n = int(input("Enter N: "))
total = 0
i = 1
print("Numbers added:", end=" ")

while i <= n:
    if i % 5 != 0:          
        if total + i > 300: 
            break
        print(i, end=" ")
        total += i
    i += 1

print("\nSum =", total)
```

## Problem 9
    Write a program that keeps accepting numbers until the user enters 0.
    Display the sum and average of all values.

```python
total = 0
count = 0

while True:
    num = float(input("Enter a number (0 to stop): "))
    if num == 0:
        break
    total += num
    count += 1

if count > 0:
    avg = total / count
    print("Sum =", total)
    print("Average =", avg)
else:
    print("No numbers were entered.")
```

## Problem 10 
    Find all numbers divisible by 7 but not multiples of 5, between 2000 and 3200 (inclusive).
    Print them as a comma-separated sequence.

```python
result = []

for i in range(2000, 3201):
    if i % 7 == 0 and i % 5 != 0:
        result.append(str(i))

for i in range(0, len(result), 10):
    print(", ".join(result[i:i+10]))
```

## Problem 11 
    Print all numbers between 1000 and 3000 (inclusive) where every digit is even.
    Output must be space-separated.

```python
result = []

for num in range(1000, 3001):
    s = str(num)
    all_even = True

    for digit in s:
        if int(digit) % 2 != 0:
            all_even = False
            break

    if all_even:
        result.append(str(num))

print(" ".join(result))
```

## Problem 12: Robot Movement
    Robot moves from (0,0) using commands:
    UP X DOWN X LEFT X RIGHT X !
    ! means stop.
    Calculate the distance from original point.
    If the result is float, print nearest integer.
    Example:
    Input: UP 5 DOWN 3 LEFT 3 RIGHT 2 !
    Output: 2

```python
x = 0
y = 0

directions = ["UP", "DOWN", "LEFT", "RIGHT"]
moves = int(input("How many moves? "))

for i in range(moves):
    direction = directions[i % 4]   # auto direction rotation
    print(f"Move {i+1}: {direction}")
    
    steps = int(input("Enter steps: "))

    if direction == "UP":
        y += steps
    elif direction == "DOWN":
        y -= steps
    elif direction == "LEFT":
        x -= steps
    elif direction == "RIGHT":
        x += steps

distance = (x**2 + y**2)**0.5
print("Distance from origin:", round(distance))
```

## Problem 13 : Prime Number Check
    Write a program to check whether a given number is prime.

```python
n = int(input("Enter a number: "))

if n <= 1:
    print("Not a prime number")
else:
    is_prime = True
    i = 2

    while i * i <= n:
        if n % i == 0:
            is_prime = False
            break
        i += 1

    if is_prime:
        print("Prime number")
    else:
        print("Not a prime number")
```

## Problem 14 : Armstrong Numbers in a Range
    Print all Armstrong numbers in a user-given range.
    (Armstrong numbers: 0, 1, 153, 370, 371, 407…)

```python
start = int(input("Enter start of range: "))
end = int(input("Enter end of range: "))

print("Armstrong numbers:")

for num in range(start, end + 1):
    digits = str(num)
    power = len(digits)
    total = sum(int(d)**power for d in digits)

    if (num == 1 or num >= 100) and total == num:
        print(num, end=" ")
```

## Problem 15 : Clock Hand Angle
    Calculate the minimum angle between hour hand and minute hand.
    Print the floor of the final angle.

    Example:
    Input: H = 9, M = 0
    Output: 90

```python
h = int(input("Enter hour: "))
m = int(input("Enter minutes: "))

h = h % 12

minute_angle = m * 6          
hour_angle = (h * 30) + (m * 0.5)   

angle = abs(hour_angle - minute_angle)

min_angle = min(angle, 360 - angle)

print(int(min_angle))   
```

## Problem 16 : Rectangle Overlap
    Given two rectangles represented by coordinates of:

    Left-top corner
    Right-bottom corner
    Check whether the two rectangles overlap.
    Rectangles sharing a side are considered overlapping.
    Assume rectangles are parallel to the axes.

```python
x1, y1 = map(float, input("Rect1 - enter x1 y1: ").split())
x2, y2 = map(float, input("Rect1 - enter x2 y2: ").split())

x3, y3 = map(float, input("Rect2 - enter x3 y3: ").split())
x4, y4 = map(float, input("Rect2 - enter x4 y4: ").split())

l1, r1 = min(x1, x2), max(x1, x2)
b1, t1 = min(y1, y2), max(y1, y2)

l2, r2 = min(x3, x4), max(x3, x4)
b2, t2 = min(y3, y4), max(y3, y4)

h_overlap = max(l1, l2) <= min(r1, r2)
v_overlap = max(b1, b2) <= min(t1, t2)

if h_overlap and v_overlap:
    print("Rectangles overlap")
else:
    print("Rectangles do not overlap")
```
